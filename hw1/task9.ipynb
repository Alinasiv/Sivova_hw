**Task 9: Спираль**
Выведите таблицу (print) размером nxn, заполненную числами от 1 до  по спирали, выходящей из левого верхнего угла и закрученной по часовой стрелке.

```python
# Сначала делаем функцию для генерациии пустой матрицы, заполненной нулями
def generate_spiral(n):
    spiral = [[0] * n for _ in range(n)]
    num = 1
# Определяем направления движения, начальные координаты и начальное направление движения
    direction = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    x, y = 0, 0
    d = 0
# Проходим по каждому элементу и задаем увеличение на 1 для каждого последующего
    for _ in range(n*n):
        spiral[x][y] = num
        num += 1
# Смотрим координаты для следующей ячейки
        dx, dy = direction[d]
        nx, ny = x + dx, y + dy
# Теперь нужно проверить, не вышло ли значение ячейки за пределы матрицы
        if 0 <= nx < n and 0 <= ny < n and spiral[nx][ny] == 0:
            x, y = nx, ny
# Иначе нужно сменить направление
        else:
            d = (d + 1) % 4 #Направление меняется циклически, то есть, если мы достигаем предела границы, то к d прибавляется 1, увеличивая индекс direction
                            #в диапазоне от 1 до 3, как только значение достигнет 4, направление возвращается к индексу 0 (вправо),
                            #что обеспечивает остаток от деления на 4
            dx, dy = direction[d]
            x, y = x + dx, y + dy

    return spiral

# Любимая консоль, надеюсь, она нужна :)
def task9(spiral):
    for row in spiral:
        print(" ".join(map(str, row)))

# Ввод размера таблицы и... готово!
n = int(input("Введите число n для создания спирали:"))
spiral = generate_spiral(n)
print("Спираль:")
task9(spiral)
```
