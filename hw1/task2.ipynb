```python
Task2: Наибольшее произведение трех чисел
Дан список, заполненный произвольными целыми числами. Найдите в этом списке ТРИ числа, произведение которых максимально.
Выведите эти числа в любом порядке. Если таких комбинаций больше 1, выведите любую.
```

```python
#Как и в задании 2, хочу вводить все в консоль. Пояснения к коду в task1
def get_numbers():
    return list(map(int, input().split()))

# Список снова должен быть отсортирован по возрастанию, пишем функцию для выявления чисел
def find_max_product_numbers(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)

# Выбираются два наименьших числа и три наибольших числа в списке по возрастанию, так как произведение двух отрицательных - положительное число
    max_product1 = sorted_numbers[n-1] * sorted_numbers[n-2] * sorted_numbers[n-3]
    max_product2 = sorted_numbers[0] * sorted_numbers[1] * sorted_numbers[n-1]

# Как было оговорено ранее, необходимы произведение и максимальных положительных, и двух минимальных отрицательных с максимальным положительным
# Если произведения равны, выводим любую из комбинаций
    if max_product1 >= max_product2:
        return [sorted_numbers[n-1], sorted_numbers[n-2], sorted_numbers[n-3]]
    else:
        return [sorted_numbers[0], sorted_numbers[1], sorted_numbers[n-1]]

# Сравниваем произведения и выдаем числа, пишем функцию второго задания
def task2():
    print("Введите числа:")
    numbers = get_numbers()
    if len(numbers) < 3:
        print('Нужно хотя бы 3 числа')
        return
    max_product_numbers = find_max_product_numbers(numbers)
    print("Числа:", *max_product_numbers)

if __name__ == "__main__":
    task2()

# Задание готово!
```
